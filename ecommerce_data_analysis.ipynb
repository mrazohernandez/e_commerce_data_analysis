```python
df = df.drop(columns=['Unnamed: 22'], errors='ignore')
```

```python
df = df.dropna(subset=['Amount'])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')
```

```python
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year
df['month_year'] = df['Date'].dt.to_period('M')
```
